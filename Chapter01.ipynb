```python
import numpy as np
import matplotlib.pyplot as plt

# Generating 10000 random numbers between 30 and 100
random_numbers = np.random.uniform(30, 100, 10000)

# Plotting the distribution of these numbers
plt.hist(random_numbers, bins=30, color='blue', edgecolor='black')
plt.title('Distribution of 10,000 student marks')
plt.xlabel('Marks')
plt.ylabel('Number of students')
plt.grid(True)
plt.savefig('Ch2uniform1.png')
plt.show()
```

```python
# Importing additional libraries for skewness and kurtosis
from scipy.stats import skew, kurtosis

# Normal Distribution
mean, std_dev = 50, 15
normal_data = np.random.normal(mean, std_dev, 10000)

# Binomial Distribution
n, p = 10, 0.5  # n = number of trials, p = probability of success in each trial
binomial_data = np.random.binomial(n, p, 10000)

# Poisson Distribution
lambda_poisson = 5  # Average number of events in a given time interval
poisson_data = np.random.poisson(lambda_poisson, 10000)

# Exponential Distribution
lambda_exp = 1.0  # Rate parameter
exponential_data = np.random.exponential(1/lambda_exp, 10000)

# Plotting all the distributions
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Normal Distribution Plot
axs[0, 0].hist(normal_data, bins=30, density=True, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Normal Distribution')
axs[0, 0].set_xlabel('Value')
axs[0, 0].set_ylabel('Frequency')

# Binomial Distribution Plot
axs[0, 1].hist(binomial_data, bins=range(n+2), align='left', color='green', edgecolor='black')
axs[0, 1].set_title('Binomial Distribution')
axs[0, 1].set_xlabel('Number of Successes')
axs[0, 1].set_ylabel('Frequency')

# Poisson Distribution Plot
axs[1, 0].hist(poisson_data, bins=30, color='purple', edgecolor='black')
axs[1, 0].set_title('Poisson Distribution')
axs[1, 0].set_xlabel('Number of Events')
axs[1, 0].set_ylabel('Frequency')

# Exponential Distribution Plot
axs[1, 1].hist(exponential_data, bins=30, color='orange', edgecolor='black')
axs[1, 1].set_title('Exponential Distribution')
axs[1, 1].set_xlabel('Time Interval')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.savefig('Ch2Distributions.png')
plt.show()

# Calculating skewness and kurtosis
skew_normal = skew(normal_data)
kurtosis_normal = kurtosis(normal_data)

# Outputs for skewness and kurtosis
skew_normal, kurtosis_normal
```

```python
# Generating data with mild skewness and kurtosis
mildly_skewed_data = np.random.beta(2, 3, 10000) * 100  # Mildly positively skewed
mildly_kurtotic_data = np.random.gamma(2, 2, 10000)  # Mild kurtosis

# Plotting Skewness with mild difference
plt.figure(figsize=(12, 6))
plt.hist(normal_data, bins=30, density=True, alpha=0.5, color='blue', label='Normal (Low Skewness)')
plt.hist(mildly_skewed_data, bins=30, density=True, alpha=0.5, color='red', label='Beta (Mild Positive Skewness)')
plt.title('Comparison of Distributions: Mild Skewness')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Calculating skewness and kurtosis for the mildly skewed and kurtotic data
skew_mildly_skewed_data = skew(mildly_skewed_data)

# Generating data with lower kurtosis
lower_kurtosis_data = np.random.normal(mean, std_dev, 10000)  # Normal distribution with a slight modification
lower_kurtosis_data = lower_kurtosis_data * np.random.uniform(0.8, 1.2, 10000)  # Introducing mild variation

# Plotting Kurtosis with lower difference
plt.figure(figsize=(12, 6))
plt.hist(normal_data, bins=30, density=True, alpha=0.5, color='blue', label='Normal (Standard Kurtosis)')
plt.hist(lower_kurtosis_data, bins=30, density=True, alpha=0.5, color='green', label='Modified Normal (Lower Kurtosis)')
plt.title('Comparison of Distributions: Mild Kurtosis')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Calculating kurtosis for the lower kurtosis data
kurtosis_lower_kurtosis_data = kurtosis(lower_kurtosis_data)

# Output for kurtosis of lower kurtosis data
kurtosis_lower_kurtosis_data
```

```python
# Combined plot for mild skewness and lower kurtosis
plt.figure(figsize=(18, 8))

# Plot for Mild Skewness
plt.subplot(1, 2, 1)
plt.hist(normal_data, bins=30, density=True, alpha=0.5, color='blue', label='Normal (Low Skewness)')
plt.hist(mildly_skewed_data, bins=30, density=True, alpha=0.5, color='red', label='Beta (Mild Positive Skewness)')
plt.title('Mild Skewness Comparison')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

# Plot for Lower Kurtosis
plt.subplot(1, 2, 2)
plt.hist(normal_data, bins=30, density=True, alpha=0.5, color='blue', label='Normal (Standard Kurtosis)')
plt.hist(lower_kurtosis_data, bins=30, density=True, alpha=0.5, color='green', label='Modified Normal (Lower Kurtosis)')
plt.title('Mild Kurtosis Comparison')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('Ch2SkewnessKurtosis.png')
plt.show()
```
