```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

# Step 1: Generate a dataset
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Features
true_slope = 2.5
true_intercept = -1.5
y = true_slope * X.squeeze() + true_intercept + np.random.randn(100) * 2  # Targets with noise

# Step 2: Fit a linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Step 3: Visualize the fit
plt.scatter(X, y, color='blue', label='Data points')
line_X = np.linspace(0, 10, 100).reshape(-1, 1)
line_y = lin_reg.predict(line_X)
plt.plot(line_X, line_y, color='red', label='Linear fit')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.title('Linear Regression Fit')
plt.show()

# Step 4: Demonstrate variance by resampling the dataset and fitting the model
for i in range(5):
    sample_indices = np.random.choice(range(100), 100, replace=True)
    X_sample = X[sample_indices]
    y_sample = y[sample_indices]
    lin_reg.fit(X_sample, y_sample)
    line_y_sample = lin_reg.predict(line_X)
    plt.plot(line_X, line_y_sample, label=f'Sample {i+1}')

plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.title('Variance in Linear Regression')
plt.show()

# Step 5: Demonstrate bias-variance tradeoff with polynomial regression
degrees = [1, 4, 15]
for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly_features.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=0)
    poly_reg = LinearRegression()
    poly_reg.fit(X_train, y_train)
    y_test_pred = poly_reg.predict(X_test)
    print(f'Degree: {degree}, MSE: {mean_squared_error(y_test, y_test_pred)}')

    # Visualize the polynomial fit
    X_range_poly = poly_features.transform(line_X)
    line_y_poly = poly_reg.predict(X_range_poly)
    plt.plot(line_X, line_y_poly, label=f'Degree {degree}')

plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.title('Bias-Variance Tradeoff with Polynomial Regression')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import ShuffleSplit

# Define a function for generating the dataset
def true_function(X):
    return np.cos(1.5 * np.pi * X)

# Generate sample data
np.random.seed(0)
n_samples = 100
X = np.sort(np.random.rand(n_samples))
y = true_function(X) + np.random.randn(n_samples) * 0.1

# Define degrees of polynomial features to be used
degrees = [1, 3, 10]

plt.figure(figsize=(14, 5))

for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)
    
    # Create a pipeline with polynomial transformation and linear regression
    polynomial_features = PolynomialFeatures(degree=degrees[i], include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    
    # Perform 10-fold cross-validation
    cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
    
    # Calculate the learning curve values
    train_sizes, train_scores, test_scores = learning_curve(pipeline, X[:, np.newaxis], y,
                                                            cv=cv, n_jobs=1, train_sizes=np.linspace(.1, 1.0, 25))
    
    # Calculate mean and standard deviation for training set scores
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    
    # Calculate mean and standard deviation for test set scores
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    # Plot the learning curve
    ax.fill_between(train_sizes, train_scores_mean - train_scores_std,
                    train_scores_mean + train_scores_std, alpha=0.1, color="r")
    ax.fill_between(train_sizes, test_scores_mean - test_scores_std,
                    test_scores_mean + test_scores_std, alpha=0.1, color="g")
    ax.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    ax.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    
    ax.set_title(f"Degree {degrees[i]}")
    ax.set_xlabel("Training examples")
    ax.set_ylabel("Score")
    ax.legend(loc="best")
    
#plt.title('Learning Curves (Bias-Variance Tradeoff)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data generation: A starry sky where each star's position is determined by linear relations and a touch of randomness
np.random.seed(0)
x = np.random.rand(100) * 10
y = 2.5 * x + np.random.randn(100) * 2

# Using the ancient method of least squares to find our line of best fit
m, b = np.polyfit(x, y, 1)

# Predicting the heavens: Calculating the y values our line predicts
y_pred = m * x + b

# Charting the cosmos: Plotting our actual data points
plt.scatter(x, y, label='Actual data', color='blue')

# Drawing the line that defines our universe's order
plt.plot(x, y_pred, color='red', label='Regression line')

# The strings of fate: Calculating and plotting the residuals
residuals = y - y_pred
for (xi, yi, ri) in zip(x, y, residuals):
    plt.plot([xi, xi], [yi, yi - ri], color='black', linestyle='--')

# Annotating our chart with the formulae of our discovered truths
plt.text(1, max(y), r'$y = mx + c$', fontsize=12, color='red')
plt.text(1, max(y) - 5, r'Residuals', fontsize=12, color='black')
plt.text(1, max(y) - 10, r'$m$: Slope', fontsize=12)
plt.text(1, max(y) - 15, r'$c$: Intercept', fontsize=12)

# Naming the axes of our charted reality
plt.xlabel('Independent variable (x)')
plt.ylabel('Dependent variable (y)')

# Legend: The key to interpreting the symbols within our celestial map
plt.legend()

# Revealing the chart to the observer
plt.savefig('Ch4Box1.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for demonstration
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Adding a column of ones to X for the intercept term
X_b = np.c_[np.ones((100, 1)), X]

# Gradient Descent Settings
learning_rate = 0.1
iterations = 10
m = len(X_b)

# Initial parameters (theta)
theta = np.random.randn(2,1)

# Function to compute the Mean Squared Error
def compute_mse(X, y, theta):
    predictions = X.dot(theta)
    errors = predictions - y
    mse = (1/m) * np.sum(errors**2)
    return mse

# Arrays to store the progress of the Gradient Descent
theta_path = [theta]
mse_path = []

# Gradient Descent Iterations
for iteration in range(iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients
    theta_path.append(theta)
    mse = compute_mse(X_b, y, theta)
    mse_path.append(mse)

# Plotting the Gradient Descent Iterations on the data
plt.figure(figsize=(10, 7))
plt.scatter(X, y)
X_new = np.array([[0], [2]])
for theta in theta_path:
    y_predict = X_b.dot(theta)
    plt.plot(X, y_predict, 'r-')

plt.xlabel("X")
plt.ylabel("y")
plt.title("Gradient Descent Iterations on Data")
plt.show()

# Correcting the mismatch in dimensions for the MSE vs Iterations plot
plt.figure(figsize=(10, 7))
plt.plot(range(iterations), mse_path, 'b-')  # Adjusting the x-axis to match the length of mse_path
plt.xlabel("Iterations")
plt.ylabel("Mean Squared Error")
plt.title("MSE vs Iterations in Gradient Descent")
plt.show()
```
